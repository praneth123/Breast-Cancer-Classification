**USING RANDOM FOREST CLASSIFIER**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load the Breast Cancer dataset
breast_cancer_dataset = load_breast_cancer()
```

```python
# Create a DataFrame from the dataset
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)

# Add the 'target' column to the DataFrame
data_frame['label'] = breast_cancer_dataset.target

```

```python
# Separate features and target
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Create a Random Forest Classifier
rf_model = RandomForestClassifier(random_state=2)

# Train the Random Forest Classifier on the training data
rf_model.fit(X_train, Y_train)

```

```python
# Evaluate accuracy on the training data
rf_train_predictions = rf_model.predict(X_train)
rf_train_accuracy = accuracy_score(Y_train, rf_train_predictions)
print('Accuracy on training data (Random Forest):', rf_train_accuracy)

```

```python
# Evaluate accuracy on the test data
rf_test_predictions = rf_model.predict(X_test)
rf_test_accuracy = accuracy_score(Y_test, rf_test_predictions)
print('Accuracy on test data (Random Forest):', rf_test_accuracy)

```

```python

# Example input data for prediction
input_data = np.array([[
    13.54, 14.36, 87.46, 566.3, 0.09779, 0.08129, 0.06664, 0.04781, 0.1885, 0.05766,
    0.2699, 0.7886, 2.058, 23.56, 0.008462, 0.0146, 0.02387, 0.01315, 0.0198, 0.0023,
    15.11, 19.26, 99.7, 711.2, 0.144, 0.1773, 0.239, 0.1288, 0.2977, 0.07259
]])

# Make predictions using the Random Forest Classifier
rf_prediction = rf_model.predict(input_data)

if rf_prediction[0] == 0:
    print('The Breast cancer is Malignant (Random Forest)')
else:
    print('The Breast Cancer is Benign (Random Forest)')
```

```python
import matplotlib.pyplot as plt

# Create a list of different numbers of trees to test
n_trees = [10, 20, 30, 40, 50]

# Initialize empty lists to store accuracy scores
train_accuracy = []
test_accuracy = []

# Train and evaluate the model for each number of trees
for n in n_trees:
    rf_model = RandomForestClassifier(n_estimators=n, random_state=2)
    rf_model.fit(X_train, Y_train)

    train_predictions = rf_model.predict(X_train)
    train_accuracy.append(accuracy_score(Y_train, train_predictions))

    test_predictions = rf_model.predict(X_test)
    test_accuracy.append(accuracy_score(Y_test, test_predictions))

# Plot the accuracy scores
plt.figure(figsize=(8, 4))
plt.plot(n_trees, train_accuracy, marker='o', label='Train Accuracy')
plt.plot(n_trees, test_accuracy, marker='o', label='Test Accuracy')
plt.xlabel('Number of Trees')
plt.ylabel('Accuracy')
plt.title('Random Forest Classifier Accuracy vs. Number of Trees')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Get feature importances from the trained model
feature_importances = rf_model.feature_importances_

# Create a DataFrame to store feature names and their importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Create a bar chart to visualize feature importances
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance in Random Forest Classifier')
plt.gca().invert_yaxis()  # Reverse the order to show the most important features at the top
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the confusion matrix for the test data
conf_matrix = confusion_matrix(Y_test, rf_test_predictions)

# Create a heatmap for the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Malignant', 'Benign'],
            yticklabels=['Malignant', 'Benign'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
